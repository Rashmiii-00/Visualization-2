```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

df1 = pd.read_csv('/content/drive/MyDrive/internet_data.csv')
df2 = pd.read_csv('/content/drive/MyDrive/churn_data.csv')
df3 = pd.read_csv('/content/drive/MyDrive/customer_data.csv')
```

```python
df = pd.merge(df1,df2, on='customerID')
```

```python
df = pd.merge(df,df3, on='customerID')
df.head()
```

```python
df = df.drop(labels=df[df['TotalCharges'] == " "].index)
```

```python
df.head()
```

```python
plt.figure(figsize=(20,8))
plt.xlabel("Monthly Charges")
plt.ylabel("Count")
sns.histplot(df, x='MonthlyCharges')
```

```python
plt.figure(figsize=(20,8))
plt.xlabel("Monthly Charges")
plt.ylabel("Count")
sns.histplot(df, x='MonthlyCharges', hue='Churn', multiple='dodge',kde=True)
```

```python
plt.figure(figsize=(20,8))
sns.histplot(df, x='tenure', hue='Churn', multiple='dodge',kde=True)
```

```python
plt.figure(figsize=(20,8))
sns.histplot(df, x='tenure', hue='gender', multiple='dodge')
```

Bar Plot: one of the parameter must be numeric
